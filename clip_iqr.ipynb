```python
from datasets import load_dataset
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = load_dataset("danjacobellis/MPQv1")
```

```python
correct = dataset['train'].filter(lambda x: x["correct"]>0.5)
incorrect = dataset['train'].filter(lambda x: x["correct"]<0.5)
plt.hist(correct['quality'],
         density=True,bins=51,
         alpha = 1,
         color='b',
         label='Correct $(\mu=0.896)$');
plt.hist(incorrect['quality'],
         density=True,
         bins=51,
         alpha = 0.7,
         color = 'r',
         label='Incorrect $(\mu=0.870)$');
plt.legend();
```

```python
high_quality = dataset['train'].filter(lambda x: x["quality"]>0.8)
hq_correct =high_quality.filter(lambda x: x["correct"]>0.5)
hq_incorrect = high_quality.filter(lambda x: x["correct"]<0.5)

plt.hist(hq_correct['quality'],
         density=True,
         bins=15,
         alpha=1,
         color='b',
         label='Correct $(\mu=0.942)$');
plt.hist(hq_incorrect['quality'],
         density=True,
         bins=15,
         alpha= 0.7,
         color='r',
         label='Incorrect $(\mu=0.938)$');
plt.legend();
```

```python
low_quality = dataset['train'].filter(lambda x: x["quality"]<0.5)
lq_correct =low_quality.filter(lambda x: x["correct"]>0.5)
lq_incorrect = low_quality.filter(lambda x: x["correct"]<0.5)

plt.hist(lq_correct['quality'],
         density=True,
         bins=15,
         label='Correct $(\mu=0.341)$',
         alpha=1,
         color='b'
        );
plt.hist(lq_incorrect['quality'],
         density=True,
         bins=15,
         alpha=0.7,
         color='r',
         label='Incorrect $(\mu=0.348)$');
plt.legend();
```
